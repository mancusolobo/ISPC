### **Ejercicio 3 :**

Calcular los siguientes límites y decir si las funciones involucradas tienen
alguna asíntota.



1.   $\displaystyle\lim_{x \to \infty}\frac{x^2-2x}{x^2+4x-4}$


2.   $\displaystyle\lim_{x \to \infty}\frac{3x^2+6x+3}{x+1}$


3.   $\displaystyle\lim_{x \to \infty}\frac{1-2x}{x}$



```python
def f1(x):
  return (x**2-2*x)/(x**2+4*x-4)
def f2(x):
  return (3*x**2+6*x+3)/(x+1)
def f3(x):
  return (1-2*x)/x
```

```python
funciones = [f1(x), f2(x), f3(x)]
for f in funciones:
  print("El limite de f" + str(funciones.index(f)+1) + " es " + str(limit(f, x, S.Infinity)))
```

```python
x = np.linspace(-5, 5, 100)
fig = plt.figure(figsize=(15,5))
ax = plt.subplot(1, 3, 1)
ax.plot(x, f1(x))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("Funcion 1")
ax = plt.subplot(1, 3, 2)
ax.plot(x, f2(x))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("Funcion 2")
ax = plt.subplot(1, 3, 3)
ax.plot(x, f3(x))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("Funcion 3")
plt.show()
```

*   La función 1 tiene asintotas vertical en sus respectivas raices que vuelven cero al denominador y vertical en $\displaystyle 1$.
$\displaystyle x = 2 (\sqrt{2}-1), x = -2 (\sqrt{2}+1)$
*   La función 2 no tiene asintota.
*   La función 3 tiene asintota horizontal en $\displaystyle 0$ y vertical en $\displaystyle -2$.





### **Ejercicio 4 :**

Calcular los siguientes límites:

1.   $\displaystyle \lim_{x \to 0} \frac{\sin x}{4x}$


2.   $\displaystyle \lim_{x \to 0} \frac{\sin 4x}{2x}$


3.   $\displaystyle\lim_{x \to \infty}(1 + \frac {2}{x})^{x}$


4.   $\displaystyle\lim_{x \to \infty}(1 + \frac {1}{x+1})^{\frac{2x+2}{3}}$

```python
def f1(x):
  return (sin(x))/(4*x)
def f2(x):
  return (sin(4*x))/(2*x)
def f3(x):
  return (1+2/x)**x
def f4(x):
  return (1+1/(x+1))**((2*x+2)/3)
```

```python
funciones = [f1(x), f2(x), f3(x), f4(x)]
limites = [0, 0, S.Infinity, S.Infinity]
for f in funciones:
  print("El limite de f" + str(funciones.index(f)+1) + " es " + str(limit(f, x, limites[funciones.index(f)])))
```
