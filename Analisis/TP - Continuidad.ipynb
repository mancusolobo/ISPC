*   $\displaystyle f\,(x)=e^x$

*   $\displaystyle c * f\,(x)$


*   $\displaystyle g\,(x) - f\,(x)$

-   $\displaystyle f\,(x)$ al ser función exponencial natural es siempre continua y creciente.

-   $\displaystyle g\,(x)$ al ser una función polinomica es siempre continua en $\displaystyle \mathbb{R}$.

-   $\displaystyle h\,(x)$ es una función continua en todo su dominio excepto en el punto donde se anula el denominador (es este caso $\displaystyle \{\mathbb{R}-0\}$).

-   Es continua por la propiedad del factor constante siempre que $\displaystyle c \in \mathbb{R}$.

-   Es continua por la propiedad de resta entre funciones continuas.

-   Una funcion racional por lo general es continua excepto en los puntos donde anulan al denominador $\displaystyle 0$ se puede decir que es discontinua evitable en $\displaystyle \mathbb{R}$.

```python
def f(x):
  return math.e**(x)
def g(x):
  return 2*x-1
def h(x):
  return 1/(x**2)
def f1(x):
  return math.random.randint(x) * f(x)
def f2(x):
  return g(x) - f(x)
def f3(x):
  return g(x)/f(x)
```

```python
x = np.linspace(-5, 5, 100)
fig, ax = plt.subplots()
ax.plot(x, f(x), label='$f(x)$')
ax.plot(x, g(x), label='$g(x)$')
ax.plot(x, h(x), label='$h(x)$')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.ylim(-20, 50)

ax.set_title("Continuidad - Ejercicio 2")
ax.legend()

plt.show()
```

### **Ejercicio 3**

Realizar una interpretación gráfica del teorema del Valor Intermedio para los
casos:

-    $\displaystyle f\,(a)<k<f\,(b)$

-    $\displaystyle f\,(b)<k<f\,(a)$

```python
def ef(x):
  return 50-10*x
funciones = [f, ef]
casos = ['$f(a)<k<f(b)$', '$f(a)>k>f(b)$']
points = {'a':1.5, 'b':3.5}
c = 2.5

x = np.linspace(-5, 5, 100)
fig = plt.figure(figsize=(20,10))

for i in range(2):
  ax = plt.subplot(1,2,i+1)

  ax.plot(x, funciones[i](x), label='f{}'.format(i+1))

  plt.ylim(-5, 50)

  ax.spines['right'].set_color('none')
  ax.spines['top'].set_color('none')
  ax.xaxis.set_ticks_position('bottom')
  ax.spines['bottom'].set_position(('data',0))
  ax.yaxis.set_ticks_position('left')
  ax.spines['left'].set_position(('data',0))

  for p in points:
    ax.plot([points[p], points[p]], [funciones[i](points[p]), 0], linestyle='--', color='darkred')
    ax.plot([points[p], 0], [funciones[i](points[p]), funciones[i](points[p])], linestyle='--', color='darkred')
    ax.scatter(points[p], funciones[i](points[p]), color='darkred')
    ax.annotate(r'{0} = {1}'.format(str(p), str(points[p])), (points[p], 0), (points[p]-0.4, -1.9), fontsize=10)
    ax.annotate(r'f({0:.2}) = {1:.2}'.format(str(p), str(funciones[i](points[p]))), (0, funciones[i](points[p])), (-1.1, funciones[i](points[p])), fontsize=10)

  ax.axhline(funciones[i](c), 0, 5, label='$k$', color='black')
  ax.plot([c, c], [funciones[i](c), 0], linestyle='--', color='darkred')
  ax.scatter(c, funciones[i](c), color='darkred')
  ax.annotate(r'{0} = {1}'.format('c', str(c)), (c, 0), (c-0.3, -1.9), fontsize=10)
  ax.annotate(r'f({0:.2}) = k = {1:.2}'.format('c', str(funciones[i](c))), (0, funciones[i](c)), (-1.5, funciones[i](c)+1.5), fontsize=10)
  ax.annotate('y=k', xy=(c, funciones[i](c)), xycoords='data', xytext=(c-0.5, f(c)+10), arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"), fontsize=10)

  ax.set_title("{}".format(casos[i]))
  ax.legend()
plt.show()
```
